# M√©todo IQR: es una medida estad√≠stica utilizada en el an√°lisis de datos para describir la dispersi√≥n o variaci√≥n.

```python
temp = pd.DataFrame()

for col in data_clean.columns:
    column = data_clean[col]
 
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)

    IQR = Q3 - Q1
    
    umbral_inferior = Q1 - 1.5 * IQR
    umbral_superior = Q3 + 1.5 * IQR

    data_clean[col + '_outliers'] = (column < umbral_inferior) | (column > umbral_superior)

    #temp[col + '_outliers'] = ((column < umbral_inferior) | (column > umbral_superior))

    print(f'{Q1 =} {Q3=} {IQR =} {umbral_inferior = } {umbral_superior =}')
```

# La transformaci√≥n de potencia se utiliza para ajustar la distribuci√≥n de los datos y puede ser √∫til en el preprocesamiento de datos para mejorar el rendimiento de ciertos modelos o an√°lisis estad√≠sticos. Utiliza la funci√≥n power_transformpara aplicando una transformaci√≥n de potencia a los datos contenidos en el DataFrame "data_noout". Luego, crea un nuevo DataFrame llamado "data_bc" que contiene los datos transformados y muestra el contenido del DataFrame resultante "data_bc"

```python
data_bc = power_transform(X=data_noout, method='box-cox')
data_bc = pd.DataFrame(data_bc, columns=data_noout.columns)
data_bc
```

```python
from sklearn.cluster import KMeans
```

# C√°lculo de la Inercia para Diferentes Valores de 'k' en K-Means

```python
inercia = []
ks = []

for k in range(1,25):
    modelo_iterado = KMeans(n_clusters = k, random_state = 5)
    modelo_iterado.fit(data_bc)
    inercia.append(modelo_iterado.inertia_)
    ks.append(k)
```

# El m√©todo del codo se utiliza para determinar el n√∫mero √≥ptimo de clusters en K-Means al analizar c√≥mo var√≠a la inercia con respecto a diferentes valores de 'k'. Este DataFrame se utiliza para visualizar y analizar la curva del codo para identificar un valor apropiado de 'k'.

```python
codo = {
    'inercia': inercia,
    'k': ks
}

codo = pd.DataFrame(codo)
```

# Este c√≥digo utiliza Seaborn para crear un gr√°fico de dispersi√≥n que muestra c√≥mo var√≠a la inercia en funci√≥n del n√∫mero de clusters 'k' y superpone una l√≠nea que conecta los puntos en el gr√°fico. 

```python
sns.scatterplot(data=codo,x='k',y='inercia',markers=True)
sns.lineplot(data=codo, x='k', y='inercia',)
```

# KMeans de Scikit-learn crear y entrenar un modelo K-Means con 5 clusters y una semilla aleatoria establecida en 10. 

```python
modelo = KMeans(n_clusters=5,random_state = 10)
modelo.fit(data_bc)
```

#  Esta columna 'cluster' contiene las etiquetas de agrupamiento asignadas a cada observaci√≥n en los datos sin valores at√≠picos.

```python
data_noout['cluster'] = modelo.labels_
data_noout
```

#  Proporcionar√° informaci√≥n sobre la distribuci√≥n de observaciones en los clusters creados por el modelo K-Means

```python
data_noout.groupby(by='cluster').count()
```

```python
sns.scatterplot(data=data_noout,x='Total_Trans_Amt', y='Total_Trans_Ct', hue='cluster')
```
